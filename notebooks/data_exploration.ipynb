Dates Analysis

```python
sales["OrderDate"] = pd.to_datetime(
    sales["OrderDate"]
)
```

```python
sales["OrderDate"].min()
```

```python
sales["OrderDate"].max()
```

```python
sales["OrderDate"].dt.year.value_counts()
```

Create Sales Amount 

```python
# SalesAmount = Quantity * Net Price 
sales["SalesAmount"] = (
    sales["Quantity"] *
    sales["Net Price"]
)
```

```python
sales["SalesAmount"].describe()
```

```python
sales.groupby(
    "Brand"
)["SalesAmount"].sum().sort_values(
    ascending=False
)
```
